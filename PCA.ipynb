```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

from IPython.display import IFrame, display
```

```python
# Centrality for citing & cited
CF = pd.DataFrame(dict(
    DEGREE = dict(C.degree),
    DEGREE_CENTRALITY = nx.degree_centrality(C),
    EIGENVECTOR = nx.eigenvector_centrality(C),
    BETWEENEESS_CENTRALITY = nx.betweenness_centrality(C),
    KATZ = nx.katz_centrality_numpy(C),
    CLOSENESS_CENTRALITY = nx.closeness_centrality(C),
    # nx.out_degree_centrality(G)
    
)) 
# Centrality for title_citing & title_cited
TF = pd.DataFrame(dict(
    DEGREE = dict(T.degree),
    DEGREE_CENTRALITY = nx.degree_centrality(T),
    EIGENVECTOR = nx.eigenvector_centrality(T),
    BETWEENEESS_CENTRALITY = nx.betweenness_centrality(T),
    KATZ = nx.katz_centrality_numpy(T),
    CLOSENESS_CENTRALITY = nx.closeness_centrality(T),
    # nx.out_degree_centrality(G)
    
)) 
# Centrality for venue_citing & venue_cited
VF = pd.DataFrame(dict(
    DEGREE = dict(V.degree),
    DEGREE_CENTRALITY = nx.degree_centrality(V),
    EIGENVECTOR = nx.eigenvector_centrality(V),
    BETWEENEESS_CENTRALITY = nx.betweenness_centrality(V),
    KATZ = nx.katz_centrality_numpy(V),
    CLOSENESS_CENTRALITY = nx.closeness_centrality(V),
    # nx.out_degree_centrality(G)
    
)) 
```
